**Print every word in upper case**

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
uppercase_words = []

for word in words:
    uppercase_words.append(word.upper())

print(uppercase_words)
```

**Create a new list containing only words with 5 or more letters**

---



```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story', 'island']
five_letters = [word for word in words if len(word) >= 5]

print(five_letters)
```



**Print the first word starting with "t"**

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story', 'island']

for word in words:
    if word.lower().startswith('t'):
        print(word)
        break
```

## Numbers


**Create a list containing the square of every number from 1 to 10**

```python
squares = [x**2 for x in range(1, 11)]
print(squares)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
odd_squares = [x**2 for x in range(1, 11) if x % 2 != 0]
print(odd_squares)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
squares_of_multiples_of_8 = [x**2 for x in range(8, 1000, 8)]
print(squares_of_multiples_of_8)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
people = [
    {"name": "Juan", "age": 34, "n_kids": 2},
    {"name": "Pepe", "age": 27, "n_kids": 0},
    {"name": "Sonia", "age": 41, "n_kids": 1},
    {"name": "Lucía", "age": 22, "n_kids": 2},
    {"name": "Leo", "age": 55, "n_kids": 5}
]

number_of_people = len(people)
print(number_of_people)
```

**How many people have kids**?

```python
people_with_kids = [person for person in people if person["n_kids"] > 0]
number_of_people_with_kids = len(people_with_kids)
print(number_of_people_with_kids)
```

**How many kids do they have in total?**

```python
total_kids = sum(person["n_kids"] for person in people if person["n_kids"] > 0)
print(total_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
people_in_a_year = [
    {
        "name": person["name"],
        "age": person["age"] + 1,  #Age = one more year
        "n_kids": person["n_kids"] + 1 if person["name"].endswith('a') else person["n_kids"]  #adding the new kid
    }
    for person in people
]

print(people_in_a_year)
```
